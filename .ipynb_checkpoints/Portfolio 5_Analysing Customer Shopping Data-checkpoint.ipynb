# Analysis of Customer Purchase Data for Dibs Retail Company

Dibs is a retail company that is fast-growing online and in Malls. They specializes in selling a wide range of products, from shoes, cosmetics, clothing to food and beverage and souvenir among other categories. Despite its rapid growth, the company has been struggling to increase its sales and customer loyalty. They have collected data on customer purchase history across different malls, but they are unsure how to use this data to their advantage. The company is interested in understanding its customers better and identifying patterns and trends in their behaviour in order to improve its sales and marketing strategies.transaction was made.

#### Definition of the columns

|Column|Meaning|
|:-----:|:-----:|
|invoice_no|Invoice number. Nominal. A combination of the letter 'I' and a 6-digit integer uniquely assigned to each operation.
|customer_id|Customer number. Nominal. A combination of the letter 'C' and a 6-digit integer uniquely assigned to each operation.
|gender|String variable of the customer's gender.
|age|Positive Integer variable of the customers age.
|category|String variable of the category of the purchased product.
|quantity|The quantities of each product (item) per transaction. Numeric.
|price|Unit price. Numeric. Product price per unit in Turkish Liras (TL).
|payment_method|String variable of the payment method (cash, credit card or debit card) used for the transaction.
|invoice_date|Invoice date. The day when a transaction was generated.
|shopping_mall|String variable of the name of the shopping mall where the transaction was made.

### Converting Data Type

```python
# Makeing a copy of the original dataset
clean_shopping_data = shopping_data.copy()
```

```python
# Converting 'gender', 'category', 'payment_method', and 'shopping_mall' to category type
clean_shopping_data["gender"] = clean_shopping_data["gender"].astype('category')
clean_shopping_data["category"] = clean_shopping_data["category"].astype('category')
clean_shopping_data["payment_method"] = clean_shopping_data["payment_method"].astype('category')
clean_shopping_data["shopping_mall"] = clean_shopping_data["shopping_mall"].astype('category')

# Converting 'invoice_date' to date-time
def convert_date(date):
    try:
        return pd.to_datetime(date, format="%m/%d/%Y", errors="coerce")
    except:
        return pd.to_datetime("1899-12-30") + pd.to_timedelta(int(date), "D")

# Applying the conversion function to the 'invoice_date' column
clean_shopping_data["invoice_date"] = convert_date(clean_shopping_data["invoice_date"])

# Verifying the data types have been correctly set
clean_shopping_data.info()
```

### Handling Missing Values

```python
# Counting the number of missing values
count_null_values = clean_shopping_data.isnull().sum()
print("Number of Null Values in the dataset: ", count_null_values)
```

```python
# Filling the missing values in 'price' with the median
clean_shopping_data["price"].fillna(clean_shopping_data["price"].median(), inplace=True)

# Filling the missing value in 'payment_method' with the most frequent value
clean_shopping_data["payment_method"].fillna(clean_shopping_data["payment_method"].mode()[0], inplace=True)

# Calculating the median date
median_date = clean_shopping_data["invoice_date"].median()

# Filling the missing values of 'invoice_date' with median
clean_shopping_data["invoice_date"].fillna(median_date, inplace=True)

# Verify that there are no more missing values
print("Number of Null Values in the dataset after replacing: ", clean_shopping_data.isnull().sum())
```

```python
# Extracting year, month, and day from 'invoice_date'
clean_shopping_data["year"] = clean_shopping_data["invoice_date"].dt.year
clean_shopping_data["month"] = clean_shopping_data["invoice_date"].dt.month
clean_shopping_data["day"] = clean_shopping_data["invoice_date"].dt.day

# Verifying the data types
clean_shopping_data.info()

# Displaying the first few rows to verify
print(clean_shopping_data[["invoice_date", "year", "month", "day"]].head())
```

```python
# Finding Q1, Q3, and Interquarlie Range of age
age_quartile_1 = clean_shopping_data["age"].quantile(0.25)
age_quartile_3 = clean_shopping_data["age"].quantile(0.75)
age_iqr = age_quartile_3 - age_quartile_1

# Setting lower and upper bound
age_lower_bound = age_quartile_1 - 1.5 * age_iqr
age_upper_bound = age_quartile_3 + 1.5 * age_iqr

# Finding the outliers
age_outliers = clean_shopping_data[(clean_shopping_data["age"] < age_lower_bound) | (clean_shopping_data["age"] > age_upper_bound)]

# Counting the number of outliers
num_age_outliers = age_outliers.shape[0]

# Printing the number of outliers
print("Number of outliers in age: ", num_age_outliers)

# Printing the outlier values for further inspection
print("Outlier values in age: ", "\n", age_outliers["age"])
```

```python
# Calculate the median for 'age'
age_median = clean_shopping_data["age"].median()

# Replace outliers in 'age' with the median
clean_shopping_data.loc[age_outliers.index, "age"] = age_median
```

```python
# Checking the outliers after fix
age_outliers_after_fix = clean_shopping_data[(clean_shopping_data["age"] < age_lower_bound) | (clean_shopping_data["age"] > age_upper_bound)]

# Counting the number of outliers
num_age_outliers_after_fix = age_outliers_after_fix.shape[0]

# Printing the number of outliers
print("Number of outliers after fixing in age: ", num_age_outliers_after_fix)
```

### Explaination
#### 1.1 Data Overview
The dataset contains 99,461 records, each representing a unique transaction of Dibs' online and shopping malls. The data includes various features such as customer demographics, transaction details, and payment methods. Key columns include invoice_no, customer_id, gender, age, category, quantity, price, payment_method, invoice_date, and shopping_mall.

#### 1.2 Data Type Conversion
- The gender, category, payment_method, and shopping_malls columns, which were initially stored as object (string), were converted to category format. This conversion is necessary for conducting categorical analysis by converting the categorical data to their respective numeric form.  
- The invoice_date column, which was initially stored as an object (string), was converted to a datetime format. This conversion is critical for conducting any time-based analysis, such as tracking sales trends over different periods.

#### 1.3 Handling Missing Values
Upon exploring the dataset, it was observed that:
- The price column has 2 missing values.
- The payment_method column has 1 missing value.
- The invoice_date column has 16 missing value.

**Actions Taken:**
- **Price:** Missing values in the price column were replaced using the median value of the price column. The median was chosen over the mean to minimize the influence of outliers.
- **Payment Method:** The missing payment_method value was replaced with the most frequent payment method, identified through mode analysis.
- **Invoice Date:** The missing invoice_date value was replaced with the median value of the invoice_date column.

#### 1.4 Handling Duplicates
The dataset was checked for duplicate records using the invoice_no as the unique identifier. There were 3 duplicate records, which were removed from the dataset for better analysis.

#### 1.5 Detecting and Handling Outliers
After plotting boxplots for age, quantity, and price to observe if outliers are present in the numeric columns, it was observed that age and price has outliers. However, after plotting boxplots for price based on all the product categories, no outlier was observed. This indicated that because of some high-price categories, the initial boxplot for price showed outliers. 

##### Actions Taken:
- **age:** There was 1 outlier present in the age column, which was replaced with the median value of age.
- **price:** Since after plotting the boxplots for price based on all the categories no outlier was not found, the outliers observed in the initial boxplot for price were not treated.

#### 1.6 Distribution Analysis
**Histograms:**
- **age:** The age distribution of customers based on gender was found to be relatively uniform.
- **quantity:** The quantity of items purchased per transaction showed a relatively uniform distribution.
- **price:** The price distributions based on each category were relatively uniform.

**Analysis:** A scatter plot was used to explore the correlation between age and spending. A positive correlation was discovered, particularly among the 30-56 age group, where spending tends to rise with age. Younger customers (18-29) and older customers (56+) tend to spend less.

## Section 3: Recommendations

```python
# Grouping sales by year and month
monthly_sales = clean_shopping_data.groupby(['year', 'month'])['price'].sum().reset_index()

# Creating a new column for 'year_month' to use in plotting
monthly_sales['year_month'] = pd.to_datetime(monthly_sales[['year', 'month']].assign(day=1))

# Plotting the monthly sales trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['year_month'], monthly_sales['price'], marker='o', color='purple')
plt.title("Monthly Sales Trend")
plt.xlabel("Year-Month")
plt.ylabel("Total Sales (TL)")
plt.grid(True)
plt.show()
```

```python
# Re-generate the colors to ensure they match the number of categories
colors = plt.cm.Paired(range(len(category_order)))

# Grouping sales by year, month, and category
monthly_sales_by_category = clean_shopping_data.groupby(['year', 'month', 'category'])['price'].sum().reset_index()

# Creating a new column for 'year_month' to use in plotting
monthly_sales_by_category['year_month'] = pd.to_datetime(monthly_sales_by_category[['year', 'month']].assign(day=1))

# Plotting the monthly sales trend for each category on the same plot using the corrected colors
plt.figure(figsize=(12, 6))
for i, category in enumerate(category_order):
    category_data = monthly_sales_by_category[monthly_sales_by_category['category'] == category]
    plt.plot(category_data['year_month'], category_data['price'], marker='o', color=colors[i], label=category)

plt.title("Monthly Sales Trend by Category")
plt.xlabel("Year-Month")
plt.ylabel("Total Sales (TL)")
plt.legend(title="Category")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Recommendations
#### 3.1 Sales by Payment Method
**Analysis:** Cash transactions were found to be the most popular payment method, followed by credit cards and debit cards. This suggests that customers prefer the traditional method of paying with cash.

**Recommendation:** Dibs should consider promoting credit card payments by providing incentives such as cashback or loyalty points. Furthermore, ensuring the security and convenience of online payment processing will entice more customers to use this method.

#### 3.2 Sales Trend Over Time
**Analysis:** The line charts revealed seasonal peaks in categories such as Clothing, Shoes, and Technology from January 2021 to January 2023, which are most likely related to major shopping events such as holidays or sales periods. However, the other categories remained almost constant, with Souvenir and Books being the most consistent over the same time period. It is very concerning that sales in all categories, with the exception of Souvenir and Books, have witnessed a drop since February 2023, with Clothing, Shoes, and Technology sales all declining significantly.

**Recommendation:** To increase sales, Dibs should introduce new collections and align marketing campaigns with peak periods. For example, increasing inventory and offering special promotions during peak sales periods could generate additional revenue. On the other hand, targeted promotions during off-peak months can aid in maintaining consistent sales.
